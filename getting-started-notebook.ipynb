# Play with Jupyter Notebooks at CourseKata Playground
CK Playground is built on technology developed as part of the [Jupyter Everywhere](https://jupytereverywhere.org/) project. CK Playground let you create, upload, edit, download, and run R or Python Jupyter notebooks completely in your browser without any need for a Jupyter hub.

## 1 Loading a notebook
You can load a Jupyter Notebook into CK Playground in three ways:
- Upload the .ipynb file directly from your hard drive using the **File > Open from file** command
- Upload the .ipynb file from a url (e.g., from a file saved on GitHub) using the **File > Open from URL** command
- Create a new R or Python notebook using the **File > New** command

## 2 Saving a notebook
You can save a notebook by downloading it as an .ipynb file. You can then upload it again when you want to continue working on it, or email it to a colleague. You also can download a notebook as a PDF file.

## 3 Sharing a notebook with a url
You can share the url by email provided the notebook is hosted on a different website (e.g., GitHub) by creating a url that, with one click, will load a notebook from a URL and open it in CK Playground. 

To do this, you simply create a link that looks like this:

> **https://play.coursekata.org/lab/index.html?from=url**

If you have a notebook stored on GitHub with this url:
> https://github.com/coursekata/teaching-materials/blob/main/By%20Chapter%20(ABCD)/Chapter%2007/InvJNB-07A-CatEV-NFL.ipynb

You can embed the url into a CourseKata Notebook link like this:
> https://play.coursekata.org/lab/index.html?from=https://github.com/coursekata/teaching-materials/blob/main/By%20Chapter%20(ABCD)/Chapter%2007/InvJNB-07A-CatEV-NFL.ipynb

If someone clicks that link, the notebook will open and be ready to use!
