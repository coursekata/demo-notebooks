### The Dataset

**Description:** The `characters` data frame contains characters from various fictional universes. At [Open Psychometrics](https://openpsychometrics.org/tests/characters/), more than 3 million volunteers from the internet rated these characters on various traits by using a sliding scale. For example, the character Mushu (from Disney's Mulan), is depicted below being rated on a scale from zero, rude, to 100, respectful. ## 4.0 - From Word Equations to Scatter Plots to Linear Models

Remember our word equation? Students start to use scatter plots to have a sense of which word equations are better than others. 

**loveable = open_minded + other stuff** is a better model than **loveable = cool + other stuff**. 

```r
# run this
gf_point(loveable ~ cool, data = characters)

gf_point(loveable ~ open_minded, data = characters, color = "forestgreen")
```

We can make precise quantitative predictions about loveable-ness as a *function* of open-mindedness. 

**loveable = f(open_mindedness) + other stuff**

**4.1:** Try modifying the "A-B-line" below to create a linear function to "fit" the data better.

```r
# modify this
gf_point(loveable ~ open_minded, data = characters, color = "forestgreen", alpha = .3) %>%
  gf_abline(intercept = 30, slope = 0)
```

**4.2:** What are some equations we came up with? What makes you think this line "fits" the data better?

From here we teach our students to understand that these functions create predictions and these predictions could be close/far from the actual data. That distance can be thought of as residual error. And we try to figure out which models both balance and minimize that error. 

```r
# just a preview for teachers
open_minded_model <- lm(loveable ~ open_minded, data = characters)

gf_point(loveable ~ open_minded, data = characters, color = "forestgreen", alpha = .3) %>%
  gf_model(open_minded_model) %>%
  gf_resid(open_minded_model)
```
